```python
# LabelEcoder

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
iris_dataset['Encoded_Species']=le.fit_transform(y.ravel())
y=le.fit_transform(y.ravel())
y
```

Decision Tree Classifier

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,y,train_size=0.19)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc_model=DecisionTreeClassifier()
```

```python
model_name=dtc_model.fit(xtrain,ytrain)
model_name
```

```python
ans2=dtc_model.predict(xtest)
ans2
```

```python
sc=dtc_model.score(xtest,ytest)
sc
```

```python
# SepalLengthCm               6.3
# SepalWidthCm                3.3
# PetalLengthCm               6.0
# PetalWidthCm                2.5
# Species          Iris-virginica -> 2
# Name: 100, dtype: object
sample = [[6.3, 3.3, 5.6, 2.5]]
ans1=dtc_model.predict(sample)
ans1
```

```python
# SepalLengthCm            5.2
# SepalWidthCm             4.1
# PetalLengthCm            1.5
# PetalWidthCm             0.1
# Species          Iris-setosa -> 0
# Name: 32, dtype: object
sample = [[5.2,4.1,1.5,0.1]]
ans2=dtc_model.predict(sample)
ans2
```

```python
# SepalLengthCm                6.4
# SepalWidthCm                 2.9
# PetalLengthCm                4.3
# PetalWidthCm                 1.3
# Species          Iris-versicolor -> 1
# Name: 74, dtype: object
sample = [[6.4,2.9,4.3,1.3]]
ans3=dtc_model.predict(sample)
ans3
```

```python
model_score_df = pd.DataFrame({'Model': model_name,
                               'Score': sc,
                               'predict on sample1': ans1,
                               'predict on sample2': ans2,
                               'predict on sample3': ans3})
model_score_df
```
